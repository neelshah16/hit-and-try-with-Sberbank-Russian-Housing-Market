```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd

x = np.arange(0,100,0.01)
```

```python
y = x**2
plt.figure(1)
plt.plot(x, y)
```

```python
y = x**0.5
plt.figure(2)
plt.plot(x, y)
```

```python
y = (x/(1+ x**0.7 ))
plt.figure(3)
plt.plot(x, y)
```

```python
y = np.log1p(x)
plt.figure(4)
plt.plot(x, y)
```

```python
y = 1/(x**0.5 + 1)
plt.figure(6)
plt.plot(x, y)
```
